```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_1 = pd.read_csv('df_1.csv')
df_2 = pd.read_csv('df_2.csv')
df_3 = pd.read_csv('df_3.csv')
```

```python
df_1_c = df_1.copy()
df_2_c = df_2.copy()
df_3_c = df_3.copy()
```

```python
df_1_c.info()
```

```python
df = pd.concat([df_1_c, df_2_c, df_3_c], axis = 0)
```

```python
lst = df.select_dtypes(include = ['number']).columns
plt.figure(figsize = (15,15))
for i, v in enumerate(lst,1):
    plt.subplot(28,3,i)
    sns.histplot(df[v], kde = True, stat = 'density')
    plt.title(f'Distribucion {v}')
    plt.xlabel(v)
    plt.ylabel('Densidad')

plt.tight_layout()
plt.show()
```

```python
matrix_corr = df[lst].corr()
plt.figure(figsize = (10,10))
plt.title('Matriz de correlacion')
sns.heatmap(matrix_corr, cmap = 'YlGnBu')
plt.show()
```

```python
matrix_corr = df[lst].corr()
plt.figure(figsize = (10,10))
plt.title('Matriz de correlacion')
sns.heatmap(matrix_corr, cmap = 'coolwarm')
plt.show()
```
